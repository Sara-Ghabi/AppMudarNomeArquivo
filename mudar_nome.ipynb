```python
import os

pasta = '/content/drive/MyDrive/Colab Notebooks/renomear_arquivo/pasta/'

#percorrendo a pasta
for nome_arquivo in os.listdir(pasta):
  #pegando caminho antigo
  caminho_antigo = os.path.join(pasta, nome_arquivo)

  #colocando o nome novo
  novo_nome = "novo_" + nome_arquivo
  caminho_novo = os.path.join(pasta, novo_nome)

  #renomeando os arquivos
  try:
    os.rename(caminho_antigo, caminho_novo)
    print(f'Renomeado de {nome_arquivo} para {novo_nome}')
  except Exception as e:
    print(f'Erro ao renomear o arquivo {nome_arquivo}: {e}')
```
